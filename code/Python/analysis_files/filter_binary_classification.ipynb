```python
# adjust the labels in merged dict so that the new merged_data reflects the filtered generics 

import json

# Load classified sentences
with open('classified_sentences_update.json', 'r') as file:
    classified_sentences = json.load(file)

# Load merged data
with open('merged_data.json', 'r') as file:
    merged_data = json.load(file)

# Create a mapping of sentences to their classification
sentence_classification = {item['sentence']: item['classification'] for item in classified_sentences}

# Update the category in merged_data based on the classification
for book in merged_data:
    for sentence_data in book:
        sentence = sentence_data['sentence']
        # Check if the sentence is in the mapping and its classification is '0'
        if sentence in sentence_classification and sentence_classification[sentence] == '0':
            sentence_data['category'] = 8  # Change category to 8

# Save the updated merged data back to the file
with open('merged_data_filtered_generics.json', 'w') as file:
    json.dump(merged_data, file, indent=4)

# Print a confirmation message
print("Updated categories in merged_data based on classifications.")
```
